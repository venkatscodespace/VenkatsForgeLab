# Line Plots - `plot()`

```python
# Importing modules
import numpy as np  # NumPy arrays boost performance in Matplotlib plots, handling large data fast
from matplotlib import pyplot as plt  # For plotting

# Array creation
x = np.arange(1, 11)  # arange(startval, endval, stepval=1) endval not included; Output: [1, ..., 11]
y = x * 2  # Output: [2, ..., 22]

# Graph 1
plt.subplot(1, 2, 1)  # plt.subplot(n_rows, n_cols, index) - First subplot in a 1x2 grid
plt.plot(x, y, color='g', linestyle=':', linewidth=2)  # Green dashed line

# Graph 2
plt.subplot(1, 2, 2)  # plt.subplot(n_rows, n_cols, index) - Second subplot in a 1x2 grid
plt.plot(x, y * 2, color='g', linestyle=':', linewidth=2)  # Green dashed line with doubled y-values

# Add titles and labels
plt.title("Hello, World!")  # Title for the plot
plt.xlabel("X")  # x-axis label
plt.ylabel("Y")  # y-axis label
plt.grid(True)  # Enabling grid for better accuracy, it is enabled for 2nd

plt.show()  # Displays the plot
```

# Bar Plots - `bar()`, `barh()`


```python
from matplotlib import pyplot as plt  # For plotting
students={"Ajay":60,"Bharath":80,"Catherine":100} #Dictionary of students and their respective marks
keys,values=list(students.keys()),list(students.values()) #converts keys and values to list
plt.bar(keys,values) #plot a bar graph

# Add titles and labels
plt.title("12A Marks")  # Title for the plot
plt.xlabel("Names")  # x-axis label
plt.ylabel("Marks")  # y-axis label
plt.grid(True)  # Enabling grid for better accuracy
plt.show() #display
```

```python
#Question: Plot a bar graph with the help of Matplotlib and a dictionary of exams and their respective marks scored by a student
from matplotlib import pyplot as plt  # For plotting
students={"Cyclic 1":60,"Cyclic 2":80,"Quarterly":100} #Dictionary of exams and their respective marks scored by AK
plt.bar(list(students.keys()),list(students.values()), color='black') #plot a vertical bar graph by converting keys and values to list

'''use plt.barh() to obtain horizontal graph'''
# Add titles and labels
plt.title("Examination Marks Comparision")  # Title for the plot
plt.xlabel("Exams ---->")  # x-axis label
plt.ylabel("Marks ---->")  # y-axis label
plt.grid(True)  # Enabling grid for better accuracy
plt.show() #display
```

# Scatter Plots - `scatter()`

```python
import numpy as np
from matplotlib import pyplot as plt  # For plotting
x=np.arange(1,11)
y=np.arange(1,11)
plt.scatter(x,y,marker="*",color='k',s=250) #s denotes size of marker point
plt.scatter(x,y*2,marker="^",color='b',s=200)
plt.title("Dummy Graph")
plt.xlabel("X Variable --->")
plt.ylabel("Y Variable --->")
plt.grid(True)
plt.show()
```

```python
#subplots
import numpy as np
from matplotlib import pyplot as plt  # For plotting
x=np.arange(1,11)
y=np.arange(1,11)

plt.subplot(2,1,1)
plt.title("Graph 1")
plt.scatter(x,y,marker=".",color='b',s=100)

plt.subplot(2,1,2)
plt.xlabel("Graph 2")
plt.scatter(x,y*2,marker=">",color='r',s=200)

plt.show()
```

# Histogram - `hist()`

```python
import numpy as np
from matplotlib import pyplot as plt
plt.hist([1,2,3,4,5,6,7,8,9],color='g')
plt.show()
```

# Pie Chart - pie()

```python
from matplotlib import pyplot as plt
fruits=['Apple','Orange','Tomato','Carrot']
quantity=[10,50,30,10]
plt.pie(quantity,labels=fruits)
plt.show()
```

# Doughnut Chart - `plt.pie([1],colors='white',radius=1)`

```python
from matplotlib import pyplot as plt
op=['I like Robert Downey Jr coming back as Dr Doom','I don\'t like Robert Downey Jr coming back as Dr Doom','I don\'t care']
q=[50,49,1]
plt.pie(q,labels=op,autopct='%0.1f%%',colors=['green','red','white'],radius=2)
plt.pie([1],colors='white',radius=1)
plt.show()
```

# Fun Facts

Here are some genuinely fun facts about `pyplot`:

1. **Easter Eggs**: Matplotlib has some hidden easter eggs. For instance, if you type `plt.style.use('ggplot')`, you can use a plotting style inspired by R's ggplot2. Its like dressing up your plots in fancy attire.

2. **Drawing with Plots**: Creative users have made art using `pyplot`. By plotting points in a specific sequence, you can draw shapes, patterns, or even complex images. There are entire competitions dedicated to creating art with code!

3. **Color Names Galore**: `pyplot` recognizes over 140 named colors. You can make your plots as colorful as a box of crayons by using names like lightcoral, darkorchid, and goldenrod.

4. **Built-in XKCD Style**: `pyplot` includes a fun `xkcd` style, which makes your plots look hand-drawn and sketchy, like the famous webcomic XKCD. Just use `plt.xkcd()` to give your charts a humorous, cartoonish feel.

5. **Plotting with Emojis**: You can use emojis as markers in your plots! Instead of using boring dots or squares, you can plot points with smileys, stars, or any emoji you like using the `marker` parameter.

6. **3D Plotting Adventures**: You can create interactive 3D plots. Rotate, zoom, and explore data in three dimensions, giving a whole new perspective to your visualizations.

7. **Animations**: `pyplot` can create animated plots, making your data dance! Whether youre visualizing changing trends over time or animating a function, it adds a dynamic element to your presentations.

8. **Geographical Maps**: Using `pyplot` with Basemap or Cartopy, you can create stunning geographical maps, plotting anything from earthquake locations to flight paths around the globe.

These fun features show that `pyplot` is not just a powerful tool for data visualization but also a playground for creativity and exploration!
